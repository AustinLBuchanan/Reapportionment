```python
p = population_2020

# state names
states = p.keys()

# total country population
p_total = sum( p[state] for state in states )

# total number of seats to distribute
k = 435

# state quotas
q = { state: k * p[state] / p_total for state in states }

print("state \t\t population \t quota")
for state in states:
    if len(state)<=6:
        print(state,"\t\t",p[state],"\t",q[state])
    else:
        print(state,"\t",p[state],"\t",q[state])
```

```python
import math

# Hamilton's method begins with the quota floors
x = { state: math.floor( q[state] ) for state in states }

# How much quota is left for each state?
rq = { state : q[state] - x[state] for state in states }

print("state \t\t so far  remaining quota")
for state in states:
    if len(state)<=6:
        print(state,"\t\t",x[state],"\t",rq[state])
    else:
        print(state,"\t",x[state],"\t",rq[state])

print( "\nTotal:", sum(x[state] for state in states), "of", k )
print( "Remaing seats to apportion:", k - sum(x[state] for state in states) )
```

```python
# function used for sorting purposes
def sort_by_second(val):
    return val[1]

states_and_remaining_quotas = [ (state, rq[state] ) for state in states ]
states_and_remaining_quotas.sort(key=sort_by_second)
```

```python
# distribute remaining seats to states with largest remaining quotas
while k > sum(x[state] for state in states):
    (state,req) = states_and_remaining_quotas.pop()
    print("Adding a seat to",state,"whose remaining quota is",req)
    x[state] += 1

print("\nstate \t\t quota \t\t\t apportionment")
for state in states:
    if len(state)<=6:
        print(state,"\t\t",q[state],"\t",x[state])
    else:
        print(state,"\t",q[state],"\t",x[state])
```
